## SECTION 29 --- Milestone Analysis
For each branch endpoint: tests which genes change significantly along that path and fits smooth expression curves.

**Parameters to set:**
- `MILESTONES_TO_ANALYZE`: copy the `branch_id` values printed by Step 29 (`review_available_branches`)
- `A_CUT`: association cutoff --- lower = more genes (noisier), higher = fewer (more selective). Default 0.3
- `P_VAL_CUT`: significance threshold for gene-pseudotime association. Default 0.001


```python
# STEP 29 — Review available branches
# Run this cell, inspect the table/figures, then choose branch IDs in Step 30.

SHOW_PLOTS = True

adata_traj, branch_table = review_available_branches(
    adata          = adata_traj,
    annotation_col = ANNOTATION_COL,
    output_dir     = selected_trajectory_dir,
    root_cluster   = ROOT_CLUSTER,
    show_plots     = SHOW_PLOTS,
)
```

```python
# STEP 30 — Analyze selected branches
# Copy branch_id values from Step 29 into MILESTONES_TO_ANALYZE.
# Leave empty until you have reviewed the branches.

MILESTONES_TO_ANALYZE = []
# Example:
# MILESTONES_TO_ANALYZE = ["7", "34"]

A_CUT     = 0.3
P_VAL_CUT = 0.001

fitted_objects = {}

if len(MILESTONES_TO_ANALYZE) == 0:
    print("No branches selected yet. Review Step 29, then fill MILESTONES_TO_ANALYZE.")
else:
    for milestone in MILESTONES_TO_ANALYZE:
        fitted_objects[str(milestone)] = run_milestone_analysis(
            adata          = adata_traj,
            milestone      = milestone,
            root_milestone = ROOT_CLUSTER,
            output_dir     = os.path.join(selected_trajectory_dir, "milestones"),
            n_jobs         = N_JOBS,
            a_cut          = A_CUT,
            p_val_cut      = P_VAL_CUT,
            name_file      = "pseudotime",  # prefix for output files
        )

    print("\nSTEP 30 COMPLETE")
```
